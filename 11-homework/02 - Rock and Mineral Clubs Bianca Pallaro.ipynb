```python
import requests

from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://rocktumbler.com/blog/rock-and-mineral-clubs/")
```

```python
#GET URLS
trs_correct = driver.find_elements_by_xpath("//tr[@bgcolor='#FFFFFF']")
#https://selenium-python.readthedocs.io/locating-elements.html#locating-by-xpath
for tr_correct in trs_correct:
    items = driver.find_elements_by_tag_name("td")
    for item in items:
        links = item.find_elements_by_tag_name('a')
        for link in links:
            url =link.get_attribute('href')
            print(url)
        
```

```python
#GET NAMES
trs_correct = driver.find_elements_by_xpath("//tr[@bgcolor='#FFFFFF']")
#https://selenium-python.readthedocs.io/locating-elements.html#locating-by-xpath
for tr_correct in trs_correct:
    items = tr_correct.find_elements_by_tag_name('td')
    for item in items:
        links = item.find_elements_by_tag_name('a')
        for link in links:
            text= link.text
            print(text)
```

```python
#GET CITIES
trs_correct = driver.find_elements_by_xpath("//tr[@bgcolor='#FFFFFF']")
#https://selenium-python.readthedocs.io/locating-elements.html#locating-by-xpath
for tr_correct in trs_correct:
    items = tr_correct.find_elements_by_tag_name('td')
    if (len(items)) == 2:
        print(items[1].text)
```

```python
#CREATE LIST OF DICTIONARIES #name, url, cities
all_rock_mineral = []
trs_correct = driver.find_elements_by_xpath("//tr[@bgcolor='#FFFFFF']")
for tr_correct in trs_correct:
    items = tr_correct.find_elements_by_tag_name('td')
    if (len(items)) == 2:
        cities = items[1].text
    for item in items:
        links = item.find_elements_by_tag_name('a')
        for link in links:
            url =link.get_attribute('href')
            text= link.text
            minerals = {
                'name': text,
                'url': url,
                'cities': cities
            }
            all_rock_mineral.append(minerals)
all_rock_mineral
```
